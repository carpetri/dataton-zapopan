```python
%pylab inline
import numpy as np
from scipy import interpolate
import pylab as py
import pandas as pd
from scipy.interpolate import Rbf
```

```python


#x, y = np.mgrid[-1:1:15j, -1:1:15j]
#fvals = func(x,y)

newfunc = interpolate.interp2d(df.long,df.lat, df.n, kind='cubic')


#Notice that we evaluate on new 1-d arrays
xnew = np.linspace(-103.5,-103.3,100)
ynew = np.linspace(20.5,21,100)

#Return the cross product
fnew = newfunc(xnew, ynew)

py.figure(1)
py.clf()
py.imshow(fnew, extent=[-103.5,-103.3,20.5,21],cmap=py.cm.jet)
```
